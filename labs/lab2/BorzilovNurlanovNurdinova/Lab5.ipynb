```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X_train = np.load('regression_x_train.npy')
y_train = np.load('regression_y_train.npy')
X_test = np.load('regression_x_test.npy')
```

```python
y = y_train.reshape(-1)
```

```python
X_train_flatten = X_train.reshape((-1, X_train.shape[2]))
```

```python
plt.imshow(y_train)
plt.show
```

```python
const = np.ones_like(y)
```

```python
X = np.concatenate((X_train_flatten, const[:, None]), axis=1)
```

```python
X[:, 35]
```

```python
from statsmodels.regression.linear_model import GLS
import statsmodels.api as sm

for i in range(35):
    simple_model = sm.GLM(y, X[:, [i, 35]]).fit()
    print( i + 1)
    print(simple_model.summary2())
    print()
```

```python
simple_model = sm.GLM(y, X_new).fit()
print(simple_model.summary2())
print()
```
